1. Write a program to segment the boat and the people on it from the background. Follow the instruction below:
- Use 'boat.jpg' as input.
- Apply Otsu thresholding.
- Draw bounding box to identify the region where the boat and people are located.

```python
img = cv.imread('boat.jpg', 0)   # 0 means grayscale

# global threshold
ret, th1 = cv.threshold(img, 127, 255, cv.THRESH_BINARY)

# Otsu's threshold
ret, th2 = cv.threshold(img, 0, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)

# Otsu's threshold after Gaussian filtering
blur = cv.GaussianBlur(img, (5, 5), 0)   # the third argument here is standard deviation
ret, th3 = cv.threshold(blur, 0, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)

# Plot grayscale images, their histogram and the resulting thresholded images
images = [img, 0, th1,
         img, 0, th2,
         blur, 0, th3]
titles = ['Original noisy image', 'Histogram', 'Global thresholding',
         'Original noisy image', 'Histogram', "Otsu's thresholding",
         'Gausssian filtered image', 'Histogram', "Otsu's thresholding"]

for i in range(3):
    plt.subplot(3, 3, i*3+1), plt.imshow(images[i*3], 'gray')
    plt.title(titles[i*3]), plt.xticks([]), plt.yticks([])
    plt.subplot(3, 3, i*3+2), plt.hist(images[i*3].ravel(), 256)
    plt.title(titles[i*3+1]), plt.xticks([]), plt.yticks([])
    plt.subplot(3, 3, i*3+3), plt.imshow(images[i*3+2], 'gray')
    plt.title(titles[i*3+2]), plt.xticks([]), plt.yticks([])
    
plt.show()
```

2. Write a program to visualize the effect of size of structuring element on the morphological transformation (e.g. erosion, dilation, opening, and closing).

- Load the 'leaf.jfif' and perform thresholding to segment the leaf from the background.
- Create a GUI containing trackbar to control the size of structuring element and display the resulting image on a window alongside the trackbar.

```python
# image 250x256 
img = cv.imread('leaf.jfif', 0)

# Trackbar function
def trackbar(x):
    ret, th1 = cv.threshold(img, x, 255, cv.THRESH_BINARY)  # 2 output ret and th1
    ret, th2 = cv.threshold(img, x, 255, cv.THRESH_BINARY_INV) 
    ret, th3 = cv.threshold(img, x, 255, cv.THRESH_TRUNC) 
    ret, th4 = cv.threshold(img, x, 255, cv.THRESH_TOZERO) 
    ret, th5 = cv.threshold(img, x, 255, cv.THRESH_TOZERO_INV) 
    
    cv.imshow('window', np.vstack([th1,th2,th3,th4,th5])) # vstack is vertically stack them argument can be list or tuple
    
cv.namedWindow('window')
cv.createTrackbar('threshold', 'window', 100 , 255, trackbar) #name of the trackbar and name of the window
cv.waitKey(0)
cv.destroyAllWindows
```

3. Apply the steps shown in exercise 4 to segment the porcelain from these images: "porcelain1.jfif" - "porcelain5.jfif".

```python
img = cv.imread('porcelain1.jfif')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img is not None

img_hsv = cv.cvtColor(img, cv.COLOR_RGB2HSV)
blue_low = (70, 20, 0)
blue_high = (120, 255, 120)

mask = cv.inRange(img_hsv, blue_low, blue_high)
res = cv.bitwise_and(img, img, mask = mask)

plt.subplot(1,2,1)
plt.imshow(mask, cmap = 'gray')
plt.subplot(1,2,2)
plt.imshow(res)
plt.show()
```
